### Basic analysis of shortened RGC experiment

```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds_z = xr.open_dataset('prog_z.nc').mean(['yh','yq'])
ds_rho0 = xr.open_dataset('prog_rho0.nc').mean(['yh','yq'])
```

```python
ds_z
```

```python
ds_z['temp'].isel(Time=0).plot()
```

```python
ds_z['temp'].isel(Time=-1).plot()
```

```python
ds_rho0['e'].isel(rho0_i=5).plot()
# plt.gca().set_ylim([-100,10])
```

```python
ds_rho0['rhopot0'].isel(Time=-1,xh=-1)
```

```python
ds_rho0['h'].cumsum('rho0_l').plot(robust=True)
plt.gca().set_ylim([1027.7,1029])
```
